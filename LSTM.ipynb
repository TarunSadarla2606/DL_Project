```python
# Import necessary libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
import torchaudio
import torchaudio.transforms as T
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
```

```python
DATA_DIR = "D:/DL_Project/meld_labels"
AUDIO_BASE_DIR = "D:/DL_Project/audio_wav" 
TRAIN_AUDIO_DIR = os.path.join(AUDIO_BASE_DIR, "train")
DEV_AUDIO_DIR = os.path.join(AUDIO_BASE_DIR, "dev") 
TEST_AUDIO_DIR = os.path.join(AUDIO_BASE_DIR, "test")

# CSV file paths
TRAIN_CSV = os.path.join(DATA_DIR, "train_sent_emo.csv")
DEV_CSV = os.path.join(DATA_DIR, "dev_sent_emo.csv")
TEST_CSV = os.path.join(DATA_DIR, "test_sent_emo.csv")

# Define emotion and sentiment mappings
EMOTIONS = {
    'neutral': 0, 
    'joy': 1, 
    'sadness': 2, 
    'anger': 3, 
    'surprise': 4, 
    'fear': 5, 
    'disgust': 6
}

SENTIMENTS = {
    'neutral': 0, 
    'positive': 1, 
    'negative': 2
}
```

```python
## 3. Data Loading and Preprocessing Functions

def get_audio_path(row, dataset_type):
    """
    Generate the correct audio path based on dataset type
    """
    file_name = f"dia{row['Dialogue_ID']}_utt{row['Utterance_ID']}.wav"
    
    if dataset_type == 'train':
        return os.path.join(TRAIN_AUDIO_DIR, file_name)
    elif dataset_type == 'dev':
        return os.path.join(DEV_AUDIO_DIR, file_name)
    elif dataset_type == 'test':
        return os.path.join(TEST_AUDIO_DIR, file_name)

def load_data(csv_path, dataset_type):
    """
    Load and preprocess data from CSV
    """
    df = pd.read_csv(csv_path)
    
    # Print basic info about the dataset
    print(f"Dataset shape: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")
    
    # Map emotions and sentiments to numerical values
    df['emotion_label'] = df['Emotion'].map(EMOTIONS)
    df['sentiment_label'] = df['Sentiment'].map(SENTIMENTS)
    
    # Generate audio file paths
    df['audio_path'] = df.apply(
        lambda row: get_audio_path(row, dataset_type), 
        axis=1
    )
    
    return df

def extract_audio_features(audio_path, max_len=300):
    """
    Extract MFCC features using torchaudio instead of librosa
    """
    try:
        # Load audio file
        waveform, sample_rate = torchaudio.load(audio_path)
        
        # Convert to mono if stereo
        if waveform.shape[0] > 1:
            waveform = torch.mean(waveform, dim=0, keepdim=True)
        
        # Extract MFCC features
        mfcc_transform = T.MFCC(
            sample_rate=sample_rate,
            n_mfcc=13,
            melkwargs={"n_fft": 400, "hop_length": 160}
        )
        mfccs = mfcc_transform(waveform)
        
        # Transpose to get time steps as first dimension
        mfccs = mfccs.squeeze().transpose(0, 1).numpy()
        
        # Pad or truncate to max_len
        if mfccs.shape[0] < max_len:
            # Pad
            pad_width = max_len - mfccs.shape[0]
            mfccs = np.pad(mfccs, ((0, pad_width), (0, 0)), mode='constant')
        else:
            # Truncate
            mfccs = mfccs[:max_len, :]
            
        return mfccs
    except Exception as e:
        print(f"Error processing {audio_path}: {e}")
        # Return zeros if there's an error
        return np.zeros((max_len, 13))
```

```python
## 4. Custom Dataset Classes

class MELDDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        
        # Count class distributions for the baseline
        emotion_counts = dataframe['emotion_label'].value_counts(normalize=True)
        sentiment_counts = dataframe['sentiment_label'].value_counts(normalize=True)
        
        self.emotion_probs = emotion_counts.sort_index().values
        self.sentiment_probs = sentiment_counts.sort_index().values
        
    def __len__(self):
        return len(self.dataframe)
    
    def __getitem__(self, idx):
        row = self.dataframe.iloc[idx]
        
        # Get audio path
        audio_path = row['audio_path']
        
        # Extract features
        if os.path.exists(audio_path):
            features = extract_audio_features(audio_path)
        else:
            print(f"Warning: Audio file not found: {audio_path}")
            features = np.zeros((300, 13))  # Placeholder
            
        # Get labels
        emotion_label = row['emotion_label']
        sentiment_label = row['sentiment_label']
        
        # Convert to tensors
        features = torch.FloatTensor(features)
        emotion_label = torch.LongTensor([emotion_label]).squeeze()
        sentiment_label = torch.LongTensor([sentiment_label]).squeeze()
        
        return features, emotion_label, sentiment_label
```

```python
## 5. Model Definitions

# Baseline Model (Random Classifier based on class distribution)
class BaselineModel:
    def __init__(self, emotion_probs, sentiment_probs):
        self.emotion_probs = emotion_probs
        self.sentiment_probs = sentiment_probs
        
    def predict_emotion(self, X):
        # Predict based on class distribution
        n_samples = len(X)
        return np.random.choice(
            len(self.emotion_probs), 
            size=n_samples, 
            p=self.emotion_probs
        )
    
    def predict_sentiment(self, X):
        # Predict based on class distribution
        n_samples = len(X)
        return np.random.choice(
            len(self.sentiment_probs), 
            size=n_samples, 
            p=self.sentiment_probs
        )
```

```python
# LSTM Model
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, emotion_classes, sentiment_classes, dropout=0.5):
        super(LSTMModel, self).__init__()
        
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        
        # LSTM layers
        self.lstm = nn.LSTM(
            input_dim, 
            hidden_dim, 
            num_layers=num_layers, 
            batch_first=True, 
            dropout=dropout if num_layers > 1 else 0,
            bidirectional=True
        )
        
        # Fully connected layers
        self.fc_shared = nn.Linear(hidden_dim * 2, hidden_dim)
        self.dropout = nn.Dropout(dropout)
        
        # Output layers for emotion and sentiment
        self.fc_emotion = nn.Linear(hidden_dim, emotion_classes)
        self.fc_sentiment = nn.Linear(hidden_dim, sentiment_classes)
        
    def forward(self, x):
        # x shape: (batch_size, seq_len, input_dim)
        
        # LSTM forward pass
        lstm_out, _ = self.lstm(x)
        
        # Take the output from the last time step
        lstm_out = lstm_out[:, -1, :]
        
        # Shared fully connected layer
        shared_features = self.fc_shared(lstm_out)
        shared_features = torch.relu(shared_features)
        shared_features = self.dropout(shared_features)
        
        # Output layers
        emotion_output = self.fc_emotion(shared_features)
        sentiment_output = self.fc_sentiment(shared_features)
        
        return emotion_output, sentiment_output
```

```python
## 6. Training and Evaluation Functions

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    """
    Train the LSTM model
    """
    # Lists to store metrics
    train_losses = []
    val_losses = []
    val_emotion_acc = []
    val_sentiment_acc = []
    
    best_val_loss = float('inf')
    best_model_weights = None
    
    for epoch in range(num_epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        
        for batch_idx, (features, emotion_labels, sentiment_labels) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Training")):
            # Move data to device
            features = features.to(device)
            emotion_labels = emotion_labels.to(device)
            sentiment_labels = sentiment_labels.to(device)
            
            # Zero the gradients
            optimizer.zero_grad()
            
            # Forward pass
            emotion_output, sentiment_output = model(features)
            
            # Calculate loss
            emotion_loss = criterion(emotion_output, emotion_labels)
            sentiment_loss = criterion(sentiment_output, sentiment_labels)
            loss = emotion_loss + sentiment_loss
            
            # Backward pass and optimize
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            
        train_loss /= len(train_loader)
        train_losses.append(train_loss)
        
        # Validation phase
        model.eval()
        val_loss = 0.0
        emotion_correct = 0
        sentiment_correct = 0
        total = 0
        
        with torch.no_grad():
            for features, emotion_labels, sentiment_labels in tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} - Validation"):
                # Move data to device
                features = features.to(device)
                emotion_labels = emotion_labels.to(device)
                sentiment_labels = sentiment_labels.to(device)
                
                # Forward pass
                emotion_output, sentiment_output = model(features)
                
                # Calculate loss
                emotion_loss = criterion(emotion_output, emotion_labels)
                sentiment_loss = criterion(sentiment_output, sentiment_labels)
                loss = emotion_loss + sentiment_loss
                
                val_loss += loss.item()
                
                # Calculate accuracy
                _, emotion_pred = torch.max(emotion_output, 1)
                _, sentiment_pred = torch.max(sentiment_output, 1)
                
                emotion_correct += (emotion_pred == emotion_labels).sum().item()
                sentiment_correct += (sentiment_pred == sentiment_labels).sum().item()
                total += emotion_labels.size(0)
        
        val_loss /= len(val_loader)
        val_losses.append(val_loss)
        
        emotion_accuracy = emotion_correct / total
        sentiment_accuracy = sentiment_correct / total
        
        val_emotion_acc.append(emotion_accuracy)
        val_sentiment_acc.append(sentiment_accuracy)
        
        print(f"Epoch {epoch+1}/{num_epochs}")
        print(f"Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")
        print(f"Val Emotion Accuracy: {emotion_accuracy:.4f}, Val Sentiment Accuracy: {sentiment_accuracy:.4f}")
        
        # Save the best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_weights = model.state_dict().copy()
    
    # Load the best model
    if best_model_weights:
        model.load_state_dict(best_model_weights)
    
    return model, {
        'train_losses': train_losses,
        'val_losses': val_losses,
        'val_emotion_acc': val_emotion_acc,
        'val_sentiment_acc': val_sentiment_acc
    }
```

```python
def evaluate_model(model, test_loader):
    """
    Evaluate the LSTM model
    """
    model.eval()
    
    true_emotions = []
    pred_emotions = []
    true_sentiments = []
    pred_sentiments = []
    
    with torch.no_grad():
        for features, emotion_labels, sentiment_labels in tqdm(test_loader, desc="Evaluating"):
            # Move data to device
            features = features.to(device)
            
            # Forward pass
            emotion_output, sentiment_output = model(features)
            
            # Get predictions
            _, emotion_preds = torch.max(emotion_output, 1)
            _, sentiment_preds = torch.max(sentiment_output, 1)
            
            # Move to CPU and convert to numpy
            true_emotions.extend(emotion_labels.cpu().numpy())
            pred_emotions.extend(emotion_preds.cpu().numpy())
            true_sentiments.extend(sentiment_labels.cpu().numpy())
            pred_sentiments.extend(sentiment_preds.cpu().numpy())
    
    # Calculate metrics
    emotion_accuracy = accuracy_score(true_emotions, pred_emotions)
    emotion_precision = precision_score(true_emotions, pred_emotions, average='weighted')
    emotion_recall = recall_score(true_emotions, pred_emotions, average='weighted')
    emotion_f1 = f1_score(true_emotions, pred_emotions, average='weighted')
    
    sentiment_accuracy = accuracy_score(true_sentiments, pred_sentiments)
    sentiment_precision = precision_score(true_sentiments, pred_sentiments, average='weighted')
    sentiment_recall = recall_score(true_sentiments, pred_sentiments, average='weighted')
    sentiment_f1 = f1_score(true_sentiments, pred_sentiments, average='weighted')
    
    # Confusion matrices
    emotion_cm = confusion_matrix(true_emotions, pred_emotions)
    sentiment_cm = confusion_matrix(true_sentiments, pred_sentiments)
    
    results = {
        'emotion': {
            'accuracy': emotion_accuracy,
            'precision': emotion_precision,
            'recall': emotion_recall,
            'f1': emotion_f1,
            'confusion_matrix': emotion_cm,
            'true_labels': true_emotions,
            'pred_labels': pred_emotions
        },
        'sentiment': {
            'accuracy': sentiment_accuracy,
            'precision': sentiment_precision,
            'recall': sentiment_recall,
            'f1': sentiment_f1,
            'confusion_matrix': sentiment_cm,
            'true_labels': true_sentiments,
            'pred_labels': pred_sentiments
        }
    }
    
    return results

def evaluate_baseline(baseline_model, test_features, true_emotions, true_sentiments):
    """
    Evaluate the baseline model
    """
    # Make predictions
    pred_emotions = baseline_model.predict_emotion(test_features)
    pred_sentiments = baseline_model.predict_sentiment(test_features)
    
    # Calculate metrics
    emotion_accuracy = accuracy_score(true_emotions, pred_emotions)
    emotion_precision = precision_score(true_emotions, pred_emotions, average='weighted')
    emotion_recall = recall_score(true_emotions, pred_emotions, average='weighted')
    emotion_f1 = f1_score(true_emotions, pred_emotions, average='weighted')
    
    sentiment_accuracy = accuracy_score(true_sentiments, pred_sentiments)
    sentiment_precision = precision_score(true_sentiments, pred_sentiments, average='weighted')
    sentiment_recall = recall_score(true_sentiments, pred_sentiments, average='weighted')
    sentiment_f1 = f1_score(true_sentiments, pred_sentiments, average='weighted')
    
    # Confusion matrices
    emotion_cm = confusion_matrix(true_emotions, pred_emotions)
    sentiment_cm = confusion_matrix(true_sentiments, pred_sentiments)
    
    results = {
        'emotion': {
            'accuracy': emotion_accuracy,
            'precision': emotion_precision,
            'recall': emotion_recall,
            'f1': emotion_f1,
            'confusion_matrix': emotion_cm
        },
        'sentiment': {
            'accuracy': sentiment_accuracy,
            'precision': sentiment_precision,
            'recall': sentiment_recall,
            'f1': sentiment_f1,
            'confusion_matrix': sentiment_cm
        }
    }
    
    return results
```

```python
def display_results(baseline_results, lstm_results):
    """
    Display evaluation results and visualizations
    """
    # Print metrics
    print("\n=== Baseline Model ===")
    print("Emotion Classification:")
    print(f"Accuracy: {baseline_results['emotion']['accuracy']:.4f}")
    print(f"Precision: {baseline_results['emotion']['precision']:.4f}")
    print(f"Recall: {baseline_results['emotion']['recall']:.4f}")
    print(f"F1-Score: {baseline_results['emotion']['f1']:.4f}")
    
    print("\nSentiment Classification:")
    print(f"Accuracy: {baseline_results['sentiment']['accuracy']:.4f}")
    print(f"Precision: {baseline_results['sentiment']['precision']:.4f}")
    print(f"Recall: {baseline_results['sentiment']['recall']:.4f}")
    print(f"F1-Score: {baseline_results['sentiment']['f1']:.4f}")
    
    print("\n=== LSTM Model ===")
    print("Emotion Classification:")
    print(f"Accuracy: {lstm_results['emotion']['accuracy']:.4f}")
    print(f"Precision: {lstm_results['emotion']['precision']:.4f}")
    print(f"Recall: {lstm_results['emotion']['recall']:.4f}")
    print(f"F1-Score: {lstm_results['emotion']['f1']:.4f}")
    
    print("\nSentiment Classification:")
    print(f"Accuracy: {lstm_results['sentiment']['accuracy']:.4f}")
    print(f"Precision: {lstm_results['sentiment']['precision']:.4f}")
    print(f"Recall: {lstm_results['sentiment']['recall']:.4f}")
    print(f"F1-Score: {lstm_results['sentiment']['f1']:.4f}")
    
    # Plot confusion matrices
    fig, axes = plt.subplots(2, 2, figsize=(20, 16))
    
    # Baseline emotion confusion matrix
    sns.heatmap(
        baseline_results['emotion']['confusion_matrix'], 
        annot=True, 
        fmt='d', 
        cmap='Blues',
        xticklabels=list(EMOTIONS.keys()),
        yticklabels=list(EMOTIONS.keys()),
        ax=axes[0, 0]
    )
    axes[0, 0].set_title('Baseline - Emotion Confusion Matrix')
    axes[0, 0].set_xlabel('Predicted')
    axes[0, 0].set_ylabel('True')
    
    # Baseline sentiment confusion matrix
    sns.heatmap(
        baseline_results['sentiment']['confusion_matrix'], 
        annot=True, 
        fmt='d', 
        cmap='Blues',
        xticklabels=list(SENTIMENTS.keys()),
        yticklabels=list(SENTIMENTS.keys()),
        ax=axes[0, 1]
    )
    axes[0, 1].set_title('Baseline - Sentiment Confusion Matrix')
    axes[0, 1].set_xlabel('Predicted')
    axes[0, 1].set_ylabel('True')
    
    # LSTM emotion confusion matrix
    sns.heatmap(
        lstm_results['emotion']['confusion_matrix'], 
        annot=True, 
        fmt='d', 
        cmap='Blues',
        xticklabels=list(EMOTIONS.keys()),
        yticklabels=list(EMOTIONS.keys()),
        ax=axes[1, 0]
    )
    axes[1, 0].set_title('LSTM - Emotion Confusion Matrix')
    axes[1, 0].set_xlabel('Predicted')
    axes[1, 0].set_ylabel('True')
    
    # LSTM sentiment confusion matrix
    sns.heatmap(
        lstm_results['sentiment']['confusion_matrix'], 
        annot=True, 
        fmt='d', 
        cmap='Blues',
        xticklabels=list(SENTIMENTS.keys()),
        yticklabels=list(SENTIMENTS.keys()),
        ax=axes[1, 1]
    )
    axes[1, 1].set_title('LSTM - Sentiment Confusion Matrix')
    axes[1, 1].set_xlabel('Predicted')
    axes[1, 1].set_ylabel('True')
    
    plt.tight_layout()
    plt.savefig('confusion_matrices.png')
    plt.show()
    
    # Plot metrics comparison
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    
    # Emotion metrics
    emotion_baseline = [
        baseline_results['emotion']['accuracy'],
        baseline_results['emotion']['precision'],
        baseline_results['emotion']['recall'],
        baseline_results['emotion']['f1']
    ]
    
    emotion_lstm = [
        lstm_results['emotion']['accuracy'],
        lstm_results['emotion']['precision'],
        lstm_results['emotion']['recall'],
        lstm_results['emotion']['f1']
    ]
    
    # Sentiment metrics
    sentiment_baseline = [
        baseline_results['sentiment']['accuracy'],
        baseline_results['sentiment']['precision'],
        baseline_results['sentiment']['recall'],
        baseline_results['sentiment']['f1']
    ]
    
    sentiment_lstm = [
        lstm_results['sentiment']['accuracy'],
        lstm_results['sentiment']['precision'],
        lstm_results['sentiment']['recall'],
        lstm_results['sentiment']['f1']
    ]
    
    fig, axes = plt.subplots(1, 2, figsize=(20, 8))
    
    # Emotion metrics plot
    x = np.arange(len(metrics))
    width = 0.35
    
    axes[0].bar(x - width/2, emotion_baseline, width, label='Baseline')
    axes[0].bar(x + width/2, emotion_lstm, width, label='LSTM')
    axes[0].set_title('Emotion Classification Metrics')
    axes[0].set_xticks(x)
    axes[0].set_xticklabels(metrics)
    axes[0].legend()
    axes[0].set_ylim(0, 1)
    
    # Sentiment metrics plot
    axes[1].bar(x - width/2, sentiment_baseline, width, label='Baseline')
    axes[1].bar(x + width/2, sentiment_lstm, width, label='LSTM')
    axes[1].set_title('Sentiment Classification Metrics')
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(metrics)
    axes[1].legend()
    axes[1].set_ylim(0, 1)
    
    plt.tight_layout()
    plt.savefig('metrics_comparison.png')
    plt.show()
```

```python
## 7. Main Execution

def main():
    """
    Main execution function
    """
    print("Loading and preprocessing data...")
    
    # Load data
    train_df = load_data(TRAIN_CSV, 'train')
    dev_df = load_data(DEV_CSV, 'dev')
    test_df = load_data(TEST_CSV, 'test')
    
    print(f"\nTrain set size: {len(train_df)}")
    print(f"Dev set size: {len(dev_df)}")
    print(f"Test set size: {len(test_df)}")
    
    # Create datasets
    train_dataset = MELDDataset(train_df)
    val_dataset = MELDDataset(dev_df)
    test_dataset = MELDDataset(test_df)
    
    # Create data loaders
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # Get class distributions for baseline model
    emotion_probs = train_dataset.emotion_probs
    sentiment_probs = train_dataset.sentiment_probs
    
    print("\nEmotion class distribution:")
    for emotion, idx in EMOTIONS.items():
        print(f"{emotion}: {emotion_probs[idx]:.4f}")
    
    print("\nSentiment class distribution:")
    for sentiment, idx in SENTIMENTS.items():
        print(f"{sentiment}: {sentiment_probs[idx]:.4f}")
    
    # Initialize baseline model
    baseline_model = BaselineModel(emotion_probs, sentiment_probs)
    
    # Initialize LSTM model
    input_dim = 13  # MFCC features
    hidden_dim = 128
    num_layers = 2
    emotion_classes = len(EMOTIONS)
    sentiment_classes = len(SENTIMENTS)
    
    lstm_model = LSTMModel(
        input_dim, 
        hidden_dim, 
        num_layers, 
        emotion_classes, 
        sentiment_classes
    ).to(device)
    
    # Define loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(lstm_model.parameters(), lr=0.001)
    
    # Train the LSTM model
    print("\nTraining LSTM model...")
    lstm_model, training_history = train_model(
        lstm_model, 
        train_loader, 
        val_loader, 
        criterion, 
        optimizer, 
        num_epochs=10
    )
    
    # Plot training history
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(training_history['train_losses'], label='Train Loss')
    plt.plot(training_history['val_losses'], label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(training_history['val_emotion_acc'], label='Emotion Accuracy')
    plt.plot(training_history['val_sentiment_acc'], label='Sentiment Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Validation Accuracy')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()
    
    # Evaluate models on test set
    print("\nEvaluating baseline model...")
    
    # Get all test features and labels for baseline
    test_features = []
    test_emotion_labels = []
    test_sentiment_labels = []
    
    for features, emotion_labels, sentiment_labels in test_loader:
        test_features.extend(features.numpy())
        test_emotion_labels.extend(emotion_labels.numpy())
        test_sentiment_labels.extend(sentiment_labels.numpy())
    
    # Evaluate baseline
    baseline_results = evaluate_baseline(
        baseline_model, 
        test_features, 
        test_emotion_labels, 
        test_sentiment_labels
    )
    
    # Evaluate LSTM
    print("\nEvaluating LSTM model...")
    lstm_results = evaluate_model(lstm_model, test_loader)
    
    # Display results
    display_results(baseline_results, lstm_results)
    
    # Save the trained model
    torch.save(lstm_model.state_dict(), 'lstm_model.pth')
    print("\nModel saved to 'lstm_model.pth'")

if __name__ == "__main__":
    main()
```
